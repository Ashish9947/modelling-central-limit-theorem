```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.stats import norm
```

```python
sample_size = 10
number_of_samples = 10000
means = []
z_scores = []
EX = 0.5      # Expected value of a uniform distribution from a to b is (a+b)/2
Var = 1/12      # Variance of a uniform distribution from a to b is ((b-a)^2)/12
for i in range(number_of_samples):
    sample = np.random.uniform(low=0, high=1, size=sample_size)
    mean = sum(sample)/sample_size
    means.append(mean)
    z_score = (mean-EX)*np.sqrt(sample_size)/np.sqrt(Var)
    z_scores.append(z_score)
```

```python
plt.xlim(-1, 1)
plt.hist(means); # As sample size --> inf, (mean-expected value) --> 0
```

```python
plt.xlim(-4, 4)
#plt.hist(means); # As sample size --> inf, (mean-expected value) --> 0
plt.hist(z_scores, density=True, bins=30);

x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x))
```
